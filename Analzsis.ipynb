```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_df = pd.read_excel("Simulated_Power_Quality_Data.xlsx")
data_df.head()
```

```python
print(data_df.columns)
```

```python
nom_voltage = 230
lim_inf = nom_voltage * 0.95
lim_sup = nom_voltage * 1.05


```

```python
print(data_df["V_L1 (V)"].describe())
print(data_df["V_L2 (V)"].describe())
print(data_df["V_L3 (V)"].describe())
```

# Voltage Analysis
The intention of this section is to present the ***voltage analysis*** of the data, under and over voltages, where in time they occur, how many times, in how much percentage

```python

plt.figure(figsize=(12, 5))
plt.plot(data_df.index, data_df["V_L1 (V)"], label="V_L1")
plt.plot(data_df.index, data_df["V_L2 (V)"], label="V_L2")
plt.plot(data_df.index, data_df["V_L3 (V)"], label="V_L3")
plt.axhline(lim_inf, color='red', linestyle='--', label=f"Lmite inferior:{lim_inf}")
plt.axhline(lim_sup, color='red', linestyle='--', label=f"Lmite superior: {lim_sup}")
plt.title("Phase Voltages")
plt.ylabel("Voltage [V]")
plt.xlabel("Time")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Power Factor Analysis

```python
plt.figure(figsize=(12, 5))
plt.plot(data_df.index, data_df["PF_L1"], label="V_L1")
plt.plot(data_df.index, data_df["PF_L2"], label="V_L2")
plt.plot(data_df.index, data_df["PF_L3"], label="V_L3")
plt.axhline(0.9, color='red', linestyle='--', label="Lmite inferior:")
plt.axhline(1, color='red', linestyle='--', label="Lmite superior:")
plt.title("Power factors")
plt.ylabel("PF [p.u.]")
plt.xlabel("Time")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Total Harmonic Distortion (THD) Analysis

```python
plt.figure(figsize=(12, 5))
plt.plot(data_df.index, data_df["THD_I_L1 (%)"], label="THD_L1",color = "blue")
plt.plot(data_df.index, data_df["THD_I_L2 (%)"], label="THD_L2",color = "black")
plt.plot(data_df.index, data_df["THD_I_L3 (%)"], label="THD_L3",color = "orange")
plt.axhline(2, color='red', linestyle='--', label="Lmite inferior:")
plt.axhline(5, color='red', linestyle='--', label="Lmite superior:")
plt.title("Phase Voltages")
plt.ylabel("Voltage [V]")
plt.xlabel("Time")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
